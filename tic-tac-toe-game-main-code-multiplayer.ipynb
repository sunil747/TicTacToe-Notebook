```python
#import for the audio and clearin screen as and when you go and run above cell
from IPython.display import Javascript,clear_output,display
#general imports
from time import sleep
#import for the color
from colorama import Fore
#for random
import random 

def startAnimation():
    '''Animations to show at the start of the game'''
    star_start = [(print('**',end=' '), sleep(0.10)) for x in range(10)]
    print(Fore.MAGENTA +'\x1b[1;31m'+'Starting Game'+'\x1b[0m'+Fore.RESET,end = ' ')
    star_end = [(print('**',end=' '), sleep(0.10)) for x in range(10)]

def take_name(player_id):
    '''Function to take the player names or randomly assign name on press enter'''

    ply1Name = ['noobExpert','inteligentIdiot','nerdTopper']
    ply2Name = ['AssumedPro','dumbLitrate','LastBenchInteligent']
    print('******press enter for auto assign******')
    name = input('enter your palyer{}Name'.format(player_id))
    #assign name based on the id of player
    if name != '':    
        pass
    else:
        if(player_id == 1):
            name = random.choice(ply1Name)
        else:
            name = random.choice(ply2Name)
    return name

def dispalyWelcomeMsg(playerName1,playerName2):
    '''Display message after entering the name'''
    print('--------------------------------------------------------------------------')
    print(Fore.GREEN+'\t\t\twelcome {}'.format(playerName1))
    print(Fore.GREEN+'\t\t\twelcome {}'.format(playerName2))
    print(Fore.RESET+'--------------------------------------------------------------------------')
    
def displayGameOptions():
    '''Show game options for the user 1 to paly 2 to know how to play and 3 to exit'''
    print("\t\tselect any option")
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    print('To play game         ('+Fore.CYAN+'enter 1'+Fore.RESET+')')
    print('To know how to play  ('+Fore.CYAN+'enter 2'+Fore.RESET+')')
    print('To exit              ('+Fore.CYAN+'enter 3'+Fore.RESET+')')
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    
def printBoard(typeOfDisplay,playerOption,playerSymbol,sq_no_lis):
    '''Display the board positions if playing or only instructions'''
    if(typeOfDisplay=='instruction'):
        sq_no_lis = [1,2,3,4,5,6,7,8,9]
        print('-----------------------')
        print('   {}   |   {}   |   {}   '.format(sq_no_lis[6],sq_no_lis[7],sq_no_lis[8]))
        print('-----------------------')
        print('   {}   |   {}   |   {}   '.format(sq_no_lis[3],sq_no_lis[4],sq_no_lis[5]))
        print('-----------------------')
        print('   {}   |   {}   |   {}   '.format(sq_no_lis[0],sq_no_lis[1],sq_no_lis[2]))
        print('-----------------------')
    elif(typeOfDisplay=="players"):
        sq_no_lis[playerOption-1] = playerSymbol
        print('-----------------------')
        print('   {}   |   {}   |   {}   '.format(sq_no_lis[6],sq_no_lis[7],sq_no_lis[8]))
        print('-----------------------')
        print('   {}   |   {}   |   {}   '.format(sq_no_lis[3],sq_no_lis[4],sq_no_lis[5]))
        print('-----------------------')
        print('   {}   |   {}   |   {}   '.format(sq_no_lis[0],sq_no_lis[1],sq_no_lis[2]))

def playGame():
    '''Main function where the game logic starts '''
    
    def getMap(indices_to_access):
        '''get the selected list positions'''
        accessed_mapping = map(sq_no_lis.__getitem__, indices_to_access)
        accessed_list = list(accessed_mapping)
        return accessed_list

    def Game_Status_Decider(sq_no_lis):
        '''main logic to decide the game winner'''
        #print('------row wise-----')
        row1 = sq_no_lis[0:3]
        row2 = sq_no_lis[3:6]
        row3 = sq_no_lis[6:9]
        #print('------row wise-----')
        #print('----column wise--------')
        indices_to_access1 = [0,3,6]
        indices_to_access2 = [1,4,7]
        indices_to_access3 = [2,5,8]
        col1 = getMap(indices_to_access1)
        col2 = getMap(indices_to_access2)
        col3 = getMap(indices_to_access3)
        #print('----column wise--------')
        #print('----diagonal wise--------')
        indices_to_access_d1 = [6,4,2]
        indices_to_access_d2 = [8,4,0]
        diag1 = getMap(indices_to_access_d1)
        diag2 = getMap(indices_to_access_d2)
        #print('----diagonal wise--------')
        if (all(elem == 'X' for elem in row1) or  all(elem == 'X' for elem in row2) or
           all(elem == 'X' for elem in row3) or all(elem == 'X' for elem in col1) or
           all(elem == 'X' for elem in col2) or all(elem == 'X' for elem in col3) or
           all(elem == 'X' for elem in diag1) or all(elem == 'X' for elem in diag2)):
            return 'player1'
        elif (all(elem == 'O' for elem in row1) or  all(elem == 'O' for elem in row2) or
           all(elem == 'O' for elem in row3) or all(elem == 'O' for elem in col1) or
           all(elem == 'O' for elem in col2) or all(elem == 'O' for elem in col3) or
           all(elem == 'O' for elem in diag1) or all(elem == 'X' for elem in diag2)):
            return 'player2'
        else:
            return 'no_winner_yet'
        
    def player_choice(player_ops):
        '''player choice to enter the respective symbol in position'''
        op_selected  = False
        while op_selected == False:
            ipn = input('enter valid position no(1,2,3,4,5,6,7,8 or 9) as your choice')
            if ipn.isdigit() == True:
                ipn = int(ipn)
                if isinstance(ipn,int) and ipn in [1,2,3,4,5,6,7,8,9]:
                    if ipn not in player_ops:
                        return ipn
                    else:
                        print("This position has alredy been entered please enter some other position")
                        continue
                else:
                    print('only numbers between 1 and 9 exist/allowed to be entered')
                    continue
                    
    clear_output()
    startAnimation()
    print('')
    #print('the board folows following pattern as number keys from the keyboard')
    print(Fore.YELLOW +'the board folows following pattern as number keys from the keyboard'+Fore.RESET)
    print(Fore.YELLOW+'---------below is the indication of the positions-------------')
    printBoard('instruction','','','')
    print('---------end of the indication of the positions-------------')
    print('player 1 follows symbol----------------> X')
    print('player 2 follows symbol----------------> O'+Fore.RESET)
    player1 = False
    player2 = True
    GameOn = True
    player_ops = []
    sq_no_lis = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    i = 0
    while GameOn:
        print()
        print()
        if Game_Status_Decider(sq_no_lis)=='no_winner_yet':
            if(player1== False):
                print('player 1 choice')
                print('-------------------------')
                pc = player_choice(player_ops)
                player_ops.append(pc)
                clear_output()
                print(Fore.YELLOW+'---------below is the indication of the positions-------------')
                printBoard('instruction','','','')
                print('---------end of the indication of the positions-------------'+Fore.RESET)
                printBoard('players',pc,'X',sq_no_lis)
                player1 = True
                player2 = False
            else:
                print('player 2 choice')
                print('-------------------------')
                pc = player_choice(player_ops)
                player_ops.append(pc)
                clear_output()
                print(Fore.YELLOW+'---------below is the indication of the positions-------------')
                printBoard('instruction','','','')
                print('---------end of the indication of the positions-------------'+Fore.RESET)
                printBoard('players',pc,'O',sq_no_lis)
                player1 = False
                player2 = True
        else:
            winner = Game_Status_Decider(sq_no_lis)
            print(Fore.RED+'-----------------------THE WINNER IS-----------------------'+Fore.RESET)
            print(winner)
            GameOn = False
            print('***************Do You Want TO Play Again???****************')
            print('-----------------------------------------------------------')
            displayGameOptions()
            make_choice()
        #condition to break while loop
        i = i+1
        if(i==9):
            GameOn = False
            print(Fore.RED+"~~~~~~~~~~~~~~GAME ENDED~~~~~~~~~~~~~~~~~")
            print("~~~~~~~~~~~~~~Bye Bye! Have a good day~~~~~~~~~~~~~~~~~"+Fore.RESET)
            
def howToPlay():
    clear_output()
    print(Fore.GREEN +'------------------>>>rules are<<<-------------------------'+ Fore.RESET)
    print(Fore.RED +'step1:'+Fore.RESET+' The board folows following pattern as number keys from the keyboard')
    printBoard('instruction','','','')
    print(Fore.RED +'step2:'+Fore.RESET+' Each No. represents the position on the board')
    print('-----------------------------------')
    print(Fore.RED +'step3:'+Fore.RESET+' Player 1 makes the first move')
    print(Fore.RED +'step4:'+Fore.RESET+' Player 1 follows symbol----------------> X')
    print(Fore.RED +'step5:'+Fore.RESET+' Player 1 chooses the position where he wants to enter the symbol X and enters respective position no')
    print('-----------------------------------')
    print(Fore.RED +'step6:'+Fore.RESET+' Player 2 makes the second move')
    print(Fore.RED +'step7:'+Fore.RESET+' Player 2 follows symbol----------------> O')
    print(Fore.RED +'step8:'+Fore.RESET+' Player 2 chooses the position where he wants to enter the symbol O and enters respective position no')
    print('-----------------------------------')
    displayGameOptions()
    print(Fore.GREEN +'vvvvvvvvvvvvvvv Enter choice Below vvvvvvvvvvvvvvvvvvvv'+ Fore.RESET)
    make_choice()
    
def toExit():
    clear_output()
    print(Fore.RED+"~~~~~~~~~~~~~~GAME ENDED~~~~~~~~~~~~~~~~~"+Fore.RESET)
    print(Fore.BLUE+"~~~~~~~~~~~~~~Bye Bye! Have a good day~~~~~~~~~~~~~~~~~"+Fore.RESET)
    
def main_method():
    def make_choice():
        '''make the choice to play or not'''
        
        def optionSelected(argument):
            '''main option in the menu'''
            dictAsSwitch = {
                1: playGame,
                2: howToPlay,
                3: toExit
            }
            # Get the function from switcher dictionary
            #.get method returns the value of the key here the value is configured to the function and there fore the func is to be executed
            func = dictAsSwitch.get(argument, lambda: "Invalid entry")
            # Execute the function
            func()
            
        op_selected  = False
        ipn = 0
        while op_selected == False:
            ipn = input('enter valid no(1,2 or3) as your choice')
            if ipn.isdigit() == True:
                ipn = int(ipn)
            if isinstance(ipn,int) and ipn>=1 and ipn<=3:
                ipn = ipn
                break
            else:
                print('only numbers between 1 and 3 exist')
                continue
        optionSelected(ipn)
    startAnimation()
    print('')
    print()
    print()
    playerName1 = take_name(1)
    playerName2 = take_name(2)
    print()
    clear_output()
    startAnimation()
    print()
    dispalyWelcomeMsg(playerName1,playerName2)
    displayGameOptions()    
    make_choice()
    
    
main_method()
```
